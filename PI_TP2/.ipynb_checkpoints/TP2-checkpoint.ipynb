**Obtenir la valeur maximale d'une image**

```python
def max_value_image(img):
    val_max = 0
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            for k in range(img.shape[2]):
                if img[i, j, k] > val_max:
                    val_max = img[i, j, k]
    return val_max

print("Intensit maximale de l'image :", max_value_image(img))
```

```python
print("Valeur maximale de l'image (numpy) :", np.max(img))
```

**Obtenir la valeur moyenne d'une image**

```python
def mean_value_image_v1(img):
    val_sum = 0
    nb_values = 0
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            for k in range(img.shape[2]):
                val_sum += img[i, j, k]
                nb_values += 1
    return val_sum / nb_values

print("Intensit moyenne de l'image (v1) :", mean_value_image_v1(img))
```

### **1. Histogramme d'intensit**

Codez la fonction histogram qui calcule l'histogramme d'intensit d'une image en niveaux de gris et affichez-le. L'histogramme d'intensit d'une image peut tre vu comme un tableau  une dimension dont chaque case contient le nombre de pixels  une certaine valeur d'intensit.

Aprs avoir cod la fonction, dcommentez la dernire ligne pour afficher l'histogramme sous forme de barres. Vous pouvez trouver un exemple de rsultat attendu dans results/histogram.png.

```python
def histogram(img):
    Image_Height = img.shape[0]
    Image_Width = img.shape[1]
    Image_color = img.shape[2]
    Historgram = np.zeros([256], np.int32)
    for x in range(0, Image_Height):
        for y in range(0, Image_Width):
            for z in range (0, Image_color):
                Historgram[img[x,y,z]] = Historgram[img[x,y,z]] + 1
    return Historgram

hist = histogram(img)
plt.bar(np.arange(len(hist)), hist)
```

### **2. Normalisation**

Codez la fonction normalize. Cette fonction a pour effet de rduire ou d'augmenter le contraste, respectivement en comprimant ou en tirant les valeurs d'intensit de l'image d'entre vers les valeurs val_min et val_max. Faites attention aux types (uint8, float...), leur dclaration tant implicite en Python.

Affichez l'image aprs normalisation avec val_min = 100 et val_max = 125. Vous pouvez trouver un exemple de rsultat attendu dans results/lena_normalized.png.

```python
def normalize(img, val_min, val_max):
    Image_Height = img.shape[0]
    Image_Width = img.shape[1]
    Image_color = img.shape[2]
    Min = np.min(img)
    Max = np.max(img)
    Normalize = img.copy()
    for x in range(0, Image_Height):
        for y in range(0, Image_Width):
            for z in range (0, Image_color):
                delta_old = Max - Min
                delta_new = val_max - val_min
                val = img[x,y,z]
                val -= Min
                val /= delta_old
                val *= delta_new
                val += val_min
                Normalize[x,y,z] = val
    return Normalize

img_normalized = normalize(img, 100, 125)
plt.imshow(img_normalized)
```

#### **Relation entre histogramme et normalisation**

Affichez  nouveau l'histogramme, cette fois-ci sur l'image normalise, et constatez les changements. Vous pouvez essayer de modifier les val_min et val_max de la fonction normalize puis d'afficher  nouveau l'histogramme. Qu'en concluez-vous ?
